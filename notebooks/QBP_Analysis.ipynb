```python
my_candidate = "Crowley, Elizabeth S"
my_candidate
```

```python
contributions_list[contributions_list.RECIPNAME == my_candidate]
```

```python
contrib_cand = contributions_list[contributions_list.RECIPNAME == my_candidate]
contrib_cand.head()
```

```python
contrib_cand.info()
```

```python
total_contrib_amnt = contrib_cand.AMNT.sum()
total_contrib_amnt
```

```python
contrib_cand.C_CODE.value_counts()
```

```python
contrib_cand.groupby("C_CODE").AMNT.sum().reset_index().sort_values("AMNT", ascending=False)
```

```python
donors_by_type = contrib_cand.fillna("").groupby(["C_CODE"]).AMNT.sum().reset_index().sort_values("AMNT", ascending=False)
donors_by_type
```

```python
donors_by_type.to_csv('../output/Crowley_Donors_by_Type.csv', index=False)
```

```python
EMPO = contrib_cand[contrib_cand.C_CODE == "EMPO"]
```

```python
EMPO_contrib_amnt = EMPO.AMNT.sum()
EMPO_contrib_amnt
```

```python
EMPO_contrib_amnt / total_contrib_amnt * 100
```
